```python
import numpy as np
from matplotlib.image import imread
import matplotlib.pyplot as plt
import os
```

```python
    A = imread('tyrion.jpg') #imported the jpg
    print(A.shape)
    plt.imshow(A)
    plt.axis('off') #x-y axis disabled
    plt.show()
```

```python
X = np.mean(A,-1) #convert it into gray-scale
plt.imshow(X).set_cmap('gray') #convert it into gray-scale
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(X, full_matrices = False) #full_matrices = False shows that we want to do the economy SVD
print(U.shape, S.shape, VT.shape)

S = np.diag(S) # S was an array of eigenvalues it is nw become a diagonal matrix 

j = 1 # Numbering the figure
for r in (5,20,100):
  # Matrix Approsimation r is the rank
  Xapprox = U[:,:r] @ S[:r,:r] @ VT[:r,:] # S[:r1,:r2] means r1 rows of S and r2 cols of S
  plt.figure(j)
  j += 1
  plt.imshow(Xapprox).set_cmap('gray')
  plt.axis('off')
  plt.title('r = ' + str(r))
  plt.show()
```
